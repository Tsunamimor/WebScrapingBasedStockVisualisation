## Define Graphing Function


In this section, we define the function `make_graph`. **You don't have to know how the function works, you should only care about the inputs. It takes a dataframe with stock data (dataframe must contain Date and Close columns), a dataframe with revenue data (dataframe must contain Date and Revenue columns), and the name of the stock.**


````python
def make_graph(stock_data, revenue_data, stock):
    # Create a subplot figure with 2 rows and 1 column
    # vertical_spacing adds space between subplots
    fig = make_subplots(rows=2, cols=1, 
                         shared_xaxes=True,  # Each subplot will have its own x-axis
                         subplot_titles=("Historical Share Price", "Historical Revenue"), 
                         vertical_spacing=0.3)

    # Remove 'infer_datetime_format=True'
    # Filter stock data to include only data up to a specific date
    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    
    # Filter revenue data to include only data up to a specific date
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']
    
    # Add share price scatter plot to the first row
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date),  # Convert dates to datetime
                              y=stock_data_specific.Close.astype("float"),  # Convert close prices to float
                              name="Share Price"), row=1, col=1)
    
    # Add revenue scatter plot to the second row
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date),  # Convert dates to datetime
                              y=revenue_data_specific.Revenue.astype("float"),  # Convert revenue to float
                              name="Revenue"), row=2, col=1)
    
    # Set x-axis labels for both subplots
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    
    # Set y-axis labels for both subplots
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    
    # Configure overall layout of the figure
    fig.update_layout(
        showlegend=False,  # Hide legend
        height=900,  # Set total height of the figure
        title=stock,  # Set title to the stock name
        xaxis_rangeslider_visible=True  # Add interactive range slider to x-axis
    )
    
    # Display the figure
    fig.show()
````

````python
!pip install requests
import requests
url ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

html_data  = requests.get(url).text
#print(html_data)
````

Parse the html data using `beautiful_soup` using parser i.e `html5lib` or `html.parser`.


````python
soup = BeautifulSoup(html_data, 'html.parser')
````

Using `BeautifulSoup` or the `read_html` function extract the table with `Tesla Revenue` and store it into a dataframe named `tesla_revenue`. The dataframe should have columns `Date` and `Revenue`.


````python
#### Here's how you can analyze all tables on a webpage and get their identifying information:
# Function to get table information
def analyze_tables(soup):
    tables = soup.find_all('table')
    table_info = []
    
    for index, table in enumerate(tables):
        info = {
            'index': index,
            'id': table.get('id', 'No ID'),
            'class': table.get('class', ['No Class']),
            'rows': len(table.find_all('tr')),
            'columns': len(table.find_all('tr')[0].find_all(['td', 'th'])) if table.find_all('tr') else 0,
            'header_text': []
        }
        
        # Try to get header text if it exists
        headers = table.find_all('th')
        if headers:
            info['header_text'] = [header.text.strip() for header in headers]
        else:
            # If no th tags, try first row
            first_row = table.find('tr')
            if first_row:
                info['header_text'] = [cell.text.strip() for cell in first_row.find_all(['td', 'th'])]
        
        table_info.append(info)
    
    return table_info

# Print the information in a readable format
def print_table_info(table_info):
    print(f"\nFound {len(table_info)} tables on the page:\n")
    
    for info in table_info:
        print(f"Table Index: {info['index']}")
        print(f"ID: {info['id']}")
        print(f"Classes: {', '.join(info['class'])}")
        print(f"Dimensions: {info['rows']} rows × {info['columns']} columns")
        if info['header_text']:
            print("Headers:")
            for header in info['header_text']:
                print(f"  - {header}")
        print("-" * 50)

# Usage example:
try:
    # Assuming you already have your soup object
    table_info = analyze_tables(soup)
    print_table_info(table_info)
    
    # You can then use this information to select the table you want
    # For example, to find a table with specific headers:
    for info in table_info:
        if 'Date' in info['header_text'] and 'Open' in info['header_text']:
            print(f"\nFound stock data table at index {info['index']}")
            # Now you can use this index to get the specific table
            target_table = soup.find_all('table')[info['index']]
            break

except Exception as e:
    print(f"An error occurred: {e}")
````

````python
### Option 1: Find table by a unique ID or class
## If the table has a specific ID like "historical-data"
#table = soup.find("table", {"id": "historical-data"})

## Or if it has a specific class
#table = soup.find("table", {"class": "stock-data-table"})

## Option 2: Find all tables and select the one we want by index
all_tables = soup.find_all("table")
#table = all_tables[desired_index]  # e.g., all_tables[0] for first table



try:
    # Find the specific table first using one of the methods above (e.g. index, class, id, etc.)
    table = all_tables[1]  # Based on the previous query above we know that the table of interest is Table Index: 1 in this case.
    
    if table is None:
        raise ValueError("Could not find the target table")
        
    tbody = table.find("tbody")
    if tbody is None:
        raise ValueError("Table body not found")
    
    for row in tbody.find_all('tr'):
        col = row.find_all("td")
        
        # Check if we have enough columns
        if len(col) < 2:
            print(f"Skipping row - insufficient columns: {len(col)}")
            continue
            
        try:
            date = col[0].text.strip()
            revenue = col[1].text.strip()
            
            tesla_revenue = pd.concat([tesla_revenue,pd.DataFrame({
                "Date":[date], 
                "Revenue":[revenue], 
            })], ignore_index=True)
        
        except IndexError as e:
            print(f"Error processing row: {e}")
            continue
            
except Exception as e:
    print(f"An error occurred: {e}")
````
